```python
import pandas as pd
import numpy as np
data = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
```

```python
# Create a barplot of the number of customers per Gender.
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
gender_plot = sns.countplot(x='Gender', data=data, palette='viridis')
plt.title('Number of Customers per Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Customers')

# Counting on top of the bars
for p in gender_plot.patches:
    gender_plot.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                         ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

```python
# Create a barplot of the number of customers per EmploymentStatus and Gender.
plt.figure(figsize=(12, 8))
employment_gender_plot = sns.countplot(x='EmploymentStatus', hue='Gender', data=data, palette='pastel')
plt.title('Number of Customers per EmploymentStatus and Gender')
plt.xlabel('EmploymentStatus')
plt.ylabel('Number of Customers')

# Display the counts on top of the bars
for p in employment_gender_plot.patches:
    employment_gender_plot.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                                    ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.legend(title='Gender')
plt.show()
```
